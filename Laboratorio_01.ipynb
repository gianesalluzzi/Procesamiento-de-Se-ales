##**Ejercicio 2**

Genere y grafique las siguiente señales discretas.


$x_1\left[n\right]=\left\{\begin{matrix}
 0 & \text{ si }   n= 0,..,20 ;\\     
                   n  & \text{ si }  n= 21,..,34
\end{matrix}\right.$

$x_2\left[n\right]=\exp \left ( \frac{n \pi}{2 } \right ) \, si \, n=4,...,39 $

Calcule su producto de convolución.


```python
#Necesitamos agregar nueva librería y sublibrería:
import math
from scipy import signal
```

```python
x1 = np.zeros(35) #arreglo 1D con 35 ceros

for i in np.arange(21,35):
  x1[i] = i

n= np.arange(4,39.0,1)
x2 = np.exp (n*math.pi/2) #Notar que se desconoce el valor de los primeros lugares (n=0,1,2,3), luego podemos hacer zero padding?

conv = signal.convolve(x1,x2) #from scipy import signal

fig, axs = plt.subplots(1,3,figsize=(16,4))
axs[0].plot(x1, color='blue', label='Señal x1')
axs[0].legend(loc='upper right')
axs[1].plot(n,x2, color='red', label='Señal x2')
axs[1].legend(loc='upper right')
axs[1].set_xlabel('n')
axs[2].plot(conv, color='g', label='Señal convolución')
axs[2].legend(loc='upper right')
axs[2].set_title('verificar la longitud de la señal')
```

##**Ejercicio 3**


Digitalice las siguientes funciones con 51 muestras (o puntos) equiespaciadas 

$f(x)= \exp(x), x \in \left[0.5,2.5 \right]$

$g(x)=sin(x), x \in \left [0, 4\pi \right ) $

```python
M=51 #jugar con la cantidad de muestras, más muestras más resolución/detalle de la señal 
x= np.linspace(0.5,2.5,M) #intervalo cerrado discretizado
f=np.exp(x)

x1= np.linspace(0,4*math.pi,M, endpoint=False)  #intervalo cerrado a izq y abierto a derecha discretizado
g=np.sin(x1)

fig, axs = plt.subplots(2,figsize=(10,8))
axs[0].plot(x,f, color='blue', label='Señal f')
axs[0].stem(x,f)
axs[0].set_xlabel('Eje x')
axs[0].set_ylabel('Eje y')
axs[0].legend() #(loc='upper right')
axs[1].plot(x1,g, color='red', label='Señal g')
axs[1].stem(x1,g)
axs[1].set_xlabel('Eje x')
axs[1].set_ylabel('Eje y')
axs[1].legend()
```

##**Ejercicio 4**

La función es $f(t)= exp(7t j)= cos(7t) + j sen(7t)$, donde $j^2=-1$ 

Por lo tanto la parte real de f es $cos(7t)$ y su parte imaginaria $sen(7t)$.

Sea la frecuencia $f=\# oscilaciones/L$, donde $L$ es la longitud del intervalo y $P=$ período. Sabemos que $P=1/f=L/\# osc$ y $f= 1/P$ 

En nuestro caso tenemos que $L=2\pi$ y la señal realiza 7 oscilaciones en el intervalo. $P= (2\pi)/7$ o bien $f= 7/(2\pi)$.

##**Ejercicio 6**

Demuestre que el conjunto de señales discretas $\left \{ s_k\left [ n \right ]=exp \left (j\frac{2\pi}{N} kn\right ) \right \}_{k=0,.., (N-1)} $. es ortogonal repecto al producto interno canónico para funciones a tiempo discreto periódicas de período $N$

```python
N=50
c1=complex(0,1)

base=np.zeros((N,N),dtype='complex')
n=np.arange(N)

for k in range(N):
  base[k]=np.exp(2*c1*np.pi/N*k*n) #k fila  y n en horizontal

#print(base)

s1=base[0] #primera fila, primera señal
s2=base[1] #segunda fila, segunda señal
s3=base[3]

print(s1,'\n')
print(s3,'\n')
print('producto interno entre s1 y s3: ', np.round(np.inner(s1,s2),decimals=3))
```

```python
basec=np.conjugate(base)
basetc=np.transpose(basec)
prod_internos= np.round(np.dot(base,basetc)/N,decimals=1) #vale para array
print( abs(prod_internos) ) #todos los Productos internos
```

```python
#graficamos la parte real de las 6 primeras señales
fig, axs = plt.subplots(6,figsize=(20,30))

for k in range(6):
  axs[k].plot(n, np.real(base[k]), color='blue')
  plt.title('Parte real de señal exp(j2pi0n/N) con k='+str(k)+' oscilaciones')
```
